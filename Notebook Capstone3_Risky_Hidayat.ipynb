#### **Metric Evaluation:**

![alt text](https://www.researchgate.net/publication/328148379/figure/fig1/AS:679514740895744@1539020347601/Model-performance-metrics-Visual-representation-of-the-classification-model-metrics.png)

- *(TP) True Positive*: *customer* `diprediksi` *cancel* dan `aktualnya` benar-benar *cancel*.
- *(FP) False Positive*: *customer* `diprediksi` cancel, tetapi sebenarnya `tidak` *cancel*.
- *(FN) False Negative*: *customer* `diprediksi` tidak *cancel*, tetapi sebenarnya *`cancel`*.
- *(TN) True Negative*: *customer* `diprediksi` tidak *cancel* dan sebenarnya `tidak` *cancel*.

Fokus pada *`False Negative`* dalam evaluasi model karena dampak ekonomi yang lebih merugikan dalam bisnis hotel. Kesalahan ini mengakibatkan `kerugian` langsung berupa kehilangan `pendapatan dan biaya` tambahan untuk mengakomodasi pembatalan yang `tidak terdeteksi`. Dengan meminimalkan yang aktualnya *cancel* namun `diprediksi` tidak cancel (`False Negative`), model dapat membantu hotel mengambil tindakan pencegahan yang tepat waktu, mengurangi kerugian, dan mempertahankan pendapatan. Pada percobaan ini kita menggunakan identifikasi sebagai berikut :

Target :
- 0 : *Not Canceled*
- 1 : *Canceled*

Tindakan hotel terhadap hasil prediksi:
Jika *customer* `diprediksi akan membatalkan`, hotel akan memberikan insentif berupa `voucher senilai 10%` untuk mendorong *customer* tetap mempertahankan pemesanan.

*Type 1 Error* : *False Positif* (*customer* yang `aktualnya` tidak *cancel* tetapi `diprediksi` *cancel*)
<br>
Konsekuensi :  Pihak hotel harus `menanggung biaya voucher` sebesar 10% dari harga hotel.

*Type 2 Error* : *False Negative* (*customer* yang `aktualnya` *cancel* tetapi `diprediksi` tidak *cancel*)
<br>
Konsekuensi : Pihak hotel harus `kehilangan kesempatan` untuk mendapat benefit sejumlah harga kamar hotel.


Untuk memberikan gambaran konsekuensi secara kuantitatif, maka kita akan coba hitung dampak biaya berdasarkan asumsi berikut :
- Rata-rata harga hotel per malam di portugal : 50 - 150 euro, kita ambil tengah yaitu 100 euro.
- Rata-rata persentase voucher gift hotel : 10% x harga hotel per malam = 10% x 100 = 10 Euro  

Dalam bisnis hotel, evaluasi fokus pada *False Negative* karena dampak ekonominya yang lebih `merugikan`, yaitu `kehilangan pendapatan` langsung dan `biaya tambahan` untuk mengakomodasi pembatalan yang `tidak terdeteksi`. Tindakan hotel terhadap hasil prediksi adalah memberikan *voucher* `senilai 10%` kepada *customer* yang `diprediksi` akan `membatalkan` untuk mendorong mereka tetap `mempertahankan pemesanan`. *False Positive* mengakibatkan pemborosan sumber daya, sedangkan *False Negative* menyebabkan hotel `melewatkan potensi pendapatan`. Dalam menghitung konsekuensi, diasumsikan harga rata-rata hotel per malam adalah 100 euro, dengan voucher hadiah senilai 10%.

### Handling Outlier

Jumlah `outlier` dalam dataset ini mencapai sekitar `28%` dari total data, yang cukup signifikan. Menghapus `outlier` dapat `menghilangkan data` dan `informasi penting` yang diperlukan untuk membangun model prediksi yang akurat. Oleh karena itu, daripada menghapusnya, kita perlu menangani outlier ini dengan metode yang tepat untuk memastikan bahwa kualitas prediksi kita tetap terjaga tanpa kehilangan informasi penting.

```python
Q1 = df['required_car_parking_spaces'].quantile(0.25)
Q3 = df['required_car_parking_spaces'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Batas Bawah: {lower_bound}")
print(f"Batas Atas: {upper_bound}")

outliers = df[(df['required_car_parking_spaces'] < lower_bound) | (df['required_car_parking_spaces'] > upper_bound)]
print("Outliers:")
outliers.head()
```

Ini menunjukkan bahwa `tidak ada` nilai di bawah atau di atas nilai 0.0 yang dianggap sebagai *outlier*. Dengan kata lain, semua nilai dalam kolom ini berada dalam `rentang yang diterima` dan `tidak dianggap sebagai *outlier*`.

```python
Q1 = df['total_of_special_requests'].quantile(0.25)
Q3 = df['total_of_special_requests'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Batas Bawah: {lower_bound}")
print(f"Batas Atas: {upper_bound}")

outliers = df[(df['total_of_special_requests'] < lower_bound) | (df['total_of_special_requests'] > upper_bound)]
print("Outliers:")
outliers.head()
```

Untuk fitur ini, `batas bawah` dihitung sebagai -1.5 dan `batas atas` dihitung sebagai 2.5. Artinya, nilai di bawah -1.5 atau di atas 2.5 akan `dianggap sebagai *outlier*`.

```python
# Fungsi untuk menghitung outlier dan memberikan batasan wajar
def compute_outliers_and_bounds(data):
    # Menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    # Menghitung Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Menghitung batas bawah dan batas atas
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menghitung outlier
    outliers = [x for x in data if x < lower_bound or x > upper_bound]

    return outliers, lower_bound, upper_bound

# Data untuk kolom-kolom yang ingin dianalisis
previous_cancellations = [0, 1, 25, 2, 3, 11, 24, 13, 6, 26, 14, 4, 19, 5, 21]
booking_changes = [0, 1, 2, 5, 3, 4, 6, 7, 9, 8, 17, 10, 13, 12, 14, 21, 16, 20, 15]
days_in_waiting_list = [0, 3, 39, 87, 77, 215, 97, 162, 46, 79, 4, 31, 122, 34, 57, 259, 50, 17, 32, 43, 75, 69, 44, 22, 113, 58, 38, 65, 48, 236, 176, 25, 108, 91, 19, 160, 68, 28, 63, 35, 42, 59, 24, 8, 183, 187, 21, 111, 56, 98, 101, 15, 16, 47, 93, 45, 14, 70, 10, 49, 178, 99, 60, 40, 85, 223, 6, 33, 391, 20, 1, 41, 147, 55, 80, 379, 150, 61, 330, 174, 107, 96, 18, 62, 13, 30, 120, 27, 207, 71, 2, 76, 125, 5, 9, 224, 53, 83, 84, 12, 73, 100, 92, 23, 11, 105, 26, 193, 7, 165, 74, 121, 36, 64, 185]

# Menghitung outlier dan batas wajar untuk kolom 'previous_cancellations'
outliers_pc, lower_bound_pc, upper_bound_pc = compute_outliers_and_bounds(previous_cancellations)
print("Outlier untuk 'previous_cancellations':", outliers_pc)
print("Batas Wajar:")
print("Batas Bawah:", lower_bound_pc)
print("Batas Atas:", upper_bound_pc)
print()

# Menghitung outlier dan batas wajar untuk kolom 'booking_changes'
outliers_bc, lower_bound_bc, upper_bound_bc = compute_outliers_and_bounds(booking_changes)
print("Outlier untuk 'booking_changes':", outliers_bc)
print("Batas Wajar:")
print("Batas Bawah:", lower_bound_bc)
print("Batas Atas:", upper_bound_bc)
print()

# Menghitung outlier dan batas wajar untuk kolom 'days_in_waiting_list'
outliers_dw, lower_bound_dw, upper_bound_dw = compute_outliers_and_bounds(days_in_waiting_list)
print("Outlier untuk 'days_in_waiting_list':", outliers_dw)
print("Batas Wajar:")
print("Batas Bawah:", lower_bound_dw)
print("Batas Atas:", upper_bound_dw)
```

## Multicolinearity

- Korelasi yang `sangat rendah` (0.089) antara **required_car_parking_spaces** dan **total_of_special_requests** menunjukkan bahwa `tidak ada` *multicollinearity* yang `signifikan` di antara fitur-fitur ini. Ini berarti bahwa fitur-fitur ini dapat digunakan bersama-sama dalam model *machine learning* tanpa khawatir tentang redundansi informasi.
- Karena fitur **reserved_room_type** `tidak memiliki variasi` dalam data, fitur ini `tidak memberikan informasi yang berguna` dan kemungkinan besar `harus dihapus` atau diproses lebih lanjut untuk analisis atau model prediktif.

```python
dfcorr=X_train.select_dtypes(include='number')
corrMatrix=dfcorr.corr(method='spearman')
already_in = set()
result = []
for col in corrMatrix:
    large_corr = [x for x in corrMatrix[col][corrMatrix[col] >= 0.6].index.tolist() if x != col]
    if large_corr and col not in already_in:
        already_in.update(set(large_corr))
        large_corr.append(col)
        result.append(large_corr)

result
```

## Pipeline

Pipeline `sangat berguna` dalam machine learning yang menyediakan struktur untuk alur kerja pemrosesan data dan pelatihan model yang `modular`, `efisien`, dan `mudah direproduksi`. Dengan pipeline, seluruh proses dari `pemrosesan data` hingga `pelatihan dan evaluasi` model dapat dilakukan dengan cara yang konsisten dan terorganisir, yang sangat penting untuk hasil yang dapat `diandalkan dan efisien`.

```python
# Urutan kategori yang diinginkan
categories1 = [['Refundable', 'No Deposit', 'Non Refund']]
categories2 = [['Low', 'High', 'Medium']]

preprocessor = ColumnTransformer(
    transformers=[
        ('Numerical 1', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', RobustScaler())
        ]), ['required_car_parking_spaces']),

        ('Numerical 2', Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
        ]), ['total_of_special_requests']),

        ('Categorical 1', Pipeline([
        ('encoder', ce.BinaryEncoder())]),['market_segment','reserved_room_type', 'region']),

        ('Categorical 2', Pipeline([
        ('encoder', OrdinalEncoder(categories=categories1))
        ]), ['deposit_type']),

        ('Categorical 3', Pipeline([
        ('encoder', OrdinalEncoder(categories=categories2))
        ]), ['previous_cancellations_segment']) ,#

        ('Categorical 4', Pipeline([
        ('encoder', OneHotEncoder(drop='first'))]), ['booking_changes_segment', 'days_in_waiting_list_segment', 'customer_type'])

    ]
)

# Definisikan pipeline tanpa classifier
pipeline = ImbPipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectKBest(score_func=f_classif)),
    ('resampling', SMOTE()),  # Oversampling dengan SMOTE
    ('classifier', RandomForestClassifier())  # Placeholder untuk classifier
])

param_grid = [
    {
        'feature_selection__k': [5, 15],  # Ubah nilai k dari 1 hingga 20
        'classifier': [CatBoostClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20]
    },
    {
        'feature_selection__k': [5, 15],  # Ubah nilai k dari 1 hingga 20
        'classifier': [RandomForestClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 4]
    },
    {
        'feature_selection__k': [5, 15],  # Ubah nilai k dari 1 hingga 20
        'classifier': [DecisionTreeClassifier()],
        'classifier__max_depth': [None, 10, 20],
        'classifier__min_samples_split': [2, 5],
        'classifier__min_samples_leaf': [1, 4]
    },
    {
        'feature_selection__k': [5, 15],  # Ubah nilai k dari 1 hingga 20
        'classifier': [LogisticRegression(max_iter=1000)],
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__solver': ['lbfgs', 'liblinear']
    },
    {
        'feature_selection__k': [5, 15],  # Ubah nilai k dari 1 hingga 20
        'classifier': [XGBClassifier()],
        'classifier__n_estimators': [50, 100, 200],
        'classifier__max_depth': [None, 10, 20],
        'classifier__learning_rate': [0.01, 0.1],
        'classifier__subsample': [0.6, 0.8]
    }
]

# Buat GridSearchCV dengan pipeline
gscv = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, scoring='f1')

# # Measure the fitting time
start_time = time.time()

# Fit GridSearchCV ke data
gscv.fit(X_train, y_train)

end_time = time.time()

fitting_time = end_time - start_time
print(f"Fitting time: {fitting_time:.2f} seconds")
```

Pipeline ini menggabungkan berbagai teknik pra-pemrosesan data untuk fitur numerik dan kategorikal, melakukan seleksi fitur, menangani ketidakseimbangan kelas dengan *SMOTE*, dan akhirnya menggunakan `XGBoost` sebagai model klasifikasi.

## **Penjelasan tentang XGBoost**

image.png

1. New Prediction:
  - Prediksi baru dihitung dengan menggabungkan prediksi sebelumnya dengan hasil pembelajaran dari residual.
2. Previous Prediction:
  - Merupakan prediksi yang dilakukan sebelumnya.
3. Learning Rate:
  - Parameter yang menentukan seberapa besar pembaruan dari residual yang akan digunakan untuk memperbaiki prediksi sebelumnya. Misalnya, pada gambar ini, learning rate (LR) adalah 0.3.
4. Residual:
  - Residual adalah selisih antara prediksi sebelumnya dan nilai aktual. Pada gambar ini, residualnya adalah -10.5.
5. Equation:
  - Persamaan yang digunakan untuk menghitung prediksi baru:
    New Prediction = Previous Prediction + (Learning rate x Residual)

    contoh pada gambar: 0.5 + (0.3 x -10.5) = - 2.65
6. The Tree made of Residual:
  - Pohon keputusan dibuat berdasarkan residual yang telah dihitung. Pohon ini digunakan untuk menentukan bagaimana residual tersebut akan digunakan untuk memperbaiki prediksi.
7. Gain:
  - Gain menunjukkan seberapa besar peningkatan akurasi prediksi setelah menambahkan pembaruan dari residual yang dihitung melalui pohon keputusan.

**Cara Kerja XGBoost:**
1. Mulai dengan prediksi awal.
2. Hitung residual sebagai perbedaan antara nilai aktual dan prediksi.
3. Buat pohon keputusan yang memprediksi residual.
4. Gunakan learning rate untuk mengontrol seberapa besar pembaruan yang akan ditambahkan ke prediksi sebelumnya.
5. Tambahkan pembaruan ini ke prediksi sebelumnya untuk mendapatkan prediksi baru.
6. Ulangi proses ini beberapa kali hingga model terkonvergensi atau mencapai jumlah iterasi yang ditentukan.
  Gambar ini secara visual menjelaskan bagaimana XGBoost secara iteratif memperbaiki prediksinya dengan menggunakan residual dan learning rate, serta bagaimana pohon keputusan digunakan untuk menangkap pola dari residual tersebut.

  - Pendapatan Hilang = Jumlah Pembatalan yang Diprediksi x Harga Rata-Rata Kamar Hotel / Malam

- **Rumus Untuk Pendapatan yang bisa Diselamatkan dengan Diskon 10%**:
  - Harga Setelah Diskon = Harga Rata-Rata Kamar Hotel / Malam x (1 - Diskon)

                         = 100 x 0.90
                         = 90 euro
  - **Pendapatan Setelah Diskon = Jumlah Pembatalan yang Diprediksi x Harga Setelah Diskon**
                              = 2.823 x 90 euro
                              = 254.070 euro

Dengan `memberikan diskon 10%` kepada customer yang diprediksi akan melakukan cancel booking:
  - Pendapatan `hilang tanpa dikson`: 282.300 euro.
  - Pendapatan `yang bisa diselamatkan dengan diskon 10%`: 254.070 euro.

`Diskon 10%` mengurangi harga rata-rata kamar menjadi 90 euro, sehingga `pendapatan yang bisa diselamatkan` dengan diskon tersebut adalah `254.070 euro`.
